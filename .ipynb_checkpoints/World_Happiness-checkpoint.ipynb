# World Happiness

The World Happiness Report is a landmark survey of the state of global happiness. 
The reports reviews the state of happiness in the world today and shows how the new science of happiness explains personal and national variations in happiness.

The happiness scores and rankings use data from the Gallup World Poll. According to the Sustainable Development Solutions Network, the scores are based on answers to the main life evaluation question asked in the poll. This question, known as the Cantril ladder, asks respondents to think of a ladder with the best possible life for them being a 10 and the worst possible life being a 0 and to rate their own current lives on that scale. 

The scores are from nationally representative samples for the years 2013-2019 and use the Gallup weights to make the estimates representative. The columns following the happiness score estimate the extent to which each of six factors – economic production, social support, life expectancy, freedom, absence of corruption, and generosity – contribute to making life evaluations higher in each country than they are in Dystopia, a hypothetical country that has values equal to the world’s lowest national averages for each of the six factors. They have no impact on the total score reported for each country, but they do explain why some countries rank higher than others.

The objective of this research is to assess:

1. What countries rank the highest in overall happiness and each of the six factors contributing to happiness?
2. What is the degree of influence and correlations of the six factors on the overall score?
3. Which factors correlate nore with the Happiness score?

For this purpose we are going to use the databases retrieved from Sustainable Development Solutions Network kaggle's database:

https://www.kaggle.com/unsdsn/world-happiness
